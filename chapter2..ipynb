```python
''' 
Variables are labels.
Objects are values.
Assignment moves the label
'''

s = "python"
s.upper()

print(s) # The original string is unchanged because strings are immutable.

s = "python"
s = s.upper()

print(s) # The variable s now references the new string "PYTHON".
# This does not raise an error. 
# You are not changing the original string in place. 
# You are creating a new uppercase string and assigning s to reference that new string.

```

Q. Regarding duck typing, What's the practical application when coding?
Duck typing means you care more about what an object can do than what exact type it is.

print_first_item([10, 20, 30])
print_first_item(("PLT", "Commercial", "Parts"))
print_first_item("Python")

Practical use: when writing reusable functions, you often do not need to force the input to be a specific type. You just need it to support the operation your function uses.


Q. Can you give me some examples of user-defined types that are either mutable or immutable?


```python
# c.sales can be changed
class Customer:
    def __init__(self, name, sales):
        self.name = name
        self.sales = sales

c = Customer("ABC Tire", 5000)
c.sales = 7000
```

```python
# An immutable-style user-defined type can be created with a frozen dataclass:
# After creation, you should not change p.sku or p.product_type.
from dataclasses import dataclass

@dataclass(frozen=True)
class Product:
    sku: str
    product_type: str

p = Product("ABC123", "PLT")
```

Q. Is "pass" same as "skip"?

Not exactly. pass means do nothing here, but keep the code syntactically valid. It is commonly used as a placeholder. 

Example:

if sales < 0:
    pass
else:
    print("Sales is valid")

continue is closer to “skip this loop iteration.”

for sales in [100, -50, 200]:
    if sales < 0:
        continue
    print(sales)

Here, -50 is skipped.

```python
values = [10, -5, 20]

for value in values:
    if value < 0:
        continue
    print(value)

# The continue statement skips the rest of the loop body for the current iteration and moves to the next iteration.
# The output will be: 10 and 20, but not -5, because -5 is less than 0 and causes the loop to skip the print statement for that iteration.
```
